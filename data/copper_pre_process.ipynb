Data Formatting

```python
# format changing
# df= df[df['status'].isin(['Won','Lost'])]
df["item_date"] = pd.to_datetime(df["item_date"], format='%Y%m%d', errors='coerce')
df["delivery date"] = pd.to_datetime(df["delivery date"],format="%Y%m%d", errors='coerce')
df["quantity tons"] = pd.to_numeric(df["quantity tons"], errors='coerce')
df["customer"] = pd.to_numeric(df["customer"], errors='coerce')
df["country"] = pd.to_numeric(df["country"], errors='coerce')
df["application"] = pd.to_numeric(df["application"], errors='coerce')
df["thickness"] = pd.to_numeric(df["thickness"], errors='coerce')
df["width"] = pd.to_numeric(df["width"], errors='coerce')
df["material_ref"] = df["material_ref"].str.lstrip('0')  # Remove leading zeros
df["product_ref"] = pd.to_numeric(df["product_ref"], errors='coerce')
df["selling_price"] = pd.to_numeric(df["selling_price"], errors='coerce')

```

```python
# Drop rows where all values are missing
df['material_ref'].fillna('unknown',inplace=True)
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
new_df = df.copy()
new_df.head(2)
```

```python
df_cleaned = new_df.dropna(subset=['id'])
df_cleaned = df_cleaned.dropna(subset=['item_date'])
df_cleaned = df_cleaned.dropna(subset=['quantity tons'])
df_cleaned = df_cleaned.dropna(subset=['customer'])
df_cleaned = df_cleaned.dropna(subset=['country'])
df_cleaned = df_cleaned.dropna(subset=['status'])
df_cleaned = df_cleaned.dropna(subset=['application'])
df_cleaned = df_cleaned.dropna(subset=['thickness'])
df_cleaned = df_cleaned.dropna(subset=['delivery date'])
df_cleaned = df_cleaned.dropna(subset=['selling_price'])
```
